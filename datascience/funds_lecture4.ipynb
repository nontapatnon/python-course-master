## Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.show()
```

### Color

```python
plt.plot(x, np.sin(x - 0), color='green') # specify color by name
plt.plot(x, np.sin(x - 1), color='b') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 3), color='#FFFF00') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color='0.25') # Grayscale between 0(black) and 1(white)
plt.plot(x, np.sin(x - 5), color=(1,0,0)) # R,G,B, each channel - between 0 and 1 
plt.show()
```

###Line Styles

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':', lw =5 ) # dotted
plt.show()
```

###Markers

```python
x = np.linspace(0, 10, 11)
plt.plot(x, x + 0, marker='o') # circle marker
plt.plot(x, x + 1, marker='.') # point marker
plt.plot(x, x + 2, marker='x') # x marker
plt.plot(x, x + 3, marker='D') # diamond marker
plt.plot(x, x + 4, marker='H') # hexagon marker
plt.plot(x, x + 5, marker='s') # square marker
plt.plot(x, x + 6, marker='+') # plus marker 
plt.plot(x, x + 7, marker='o',markerfacecolor = 'g', markersize = 12) 
plt.show()
```

###Formatting String

```python
plt.plot(x, x + 0, '-c') # solid, cyan
plt.plot(x, x + 1, '--m') # dashed, magenta
plt.plot(x, x + 2, '-y.') # dashdot, yellow, circle marker
plt.plot(x, x + 3, 'k:+'); # black, dotted, plus marker
plt.show()
```

###Labeling Plots

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.title("A Sine and Cosine Curve")
plt.xlabel("x")
plt.ylabel("f(x)")
#plt.legend()
plt.legend(loc=1)
plt.show()
```

##Object-oriented Style

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_axes([0, 0, 1, 1]) # main axes
#ax = plt.axes() # no need for figure object
ax.plot(x, np.sin(x))
ax.set_xlabel('angle')
ax.set_title('sin')
ax.set_xticks([0,2,4,6,8,10])
ax.set_xticklabels(['zero','two','four','six','eight','ten'])
ax.set_yticks([-1,0,1])
plt.show()
```

```python
from matplotlib import pyplot as plt

ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
xlabel='x', ylabel='sin(x)',
title='A Simple Plot')
plt.show()
```

```python
rng = np.random.RandomState(0) #reproducable results
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis')
plt.colorbar(); # show color scale
```

#Subplot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 60)
plt.subplot(3, 1, 1, frameon=False) 
#plt.subplot(311, frameon=False)  
plt.plot(x, np.sin(x), 'o-')
plt.title('A tale of 3 subplots')
plt.ylabel('sin(x')
plt.subplot(3, 1, 3, facecolor='lightgreen')
plt.plot(x, np.cos(x), '--')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```
